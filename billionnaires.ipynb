<h3>Data Preparation for Question 1.</h3><br>
Which industries produce the highest number of billionaires, and how does their average wealth compare across different industries? <br>
Creating a new dataframe (IndustryWealthCorelation) for number of billionnaires by industry, and the average wealth for each industry.

```python
# Grouping by 'industries' and calculating the number of billionaires and their average wealth
IndustryWealthCorrelation = (
    billionnaires.groupby("industries")["finalWorth"]
    .agg(["count", "mean"])
    .reset_index()
    .rename(columns={"industries": "Industry", "count": "Number_of_Billionaires", "mean": "Average_Wealth"})
)

# Sorting the dataframe by Number_of_Billionaires in descending order
IndustryWealthCorrelation = IndustryWealthCorrelation.sort_values(by="Number_of_Billionaires", ascending=False)

# Display the resulting dataframe

print(IndustryWealthCorrelation.head(20))
print(IndustryWealthCorrelation.describe())

print('The industry with the highest number of billionaires is {} with {} billionaires.'.format(IndustryWealthCorrelation.iloc[0, 0], IndustryWealthCorrelation.iloc[0, 1]))


# Creating a new dataframe sorted by Average Wealth in descending order
IndustryWealthCorrelation_AverageWealth = IndustryWealthCorrelation.sort_values(by="Average_Wealth", ascending=False)

print('The industry with the highest average wealth is the {} industry with an average wealth of ${:,.2f} US Dollars.'.format(IndustryWealthCorrelation_AverageWealth.iloc[0, 0], IndustryWealthCorrelation_AverageWealth.iloc[0, 2]*1000000))
```

<h1>Data Preparation for Question 3</h1><br>
What are the key differences in industry, geographic distribution, and final net worth between self-made billionaires and those with inherited wealth?<br>

Creating a new dataframe with industry (industries), net worth (finalworth), self-made (selfMade), GDP (gdp_country), CPI (cpi_country) and tax revenue (tax_revenue_country_country).


```python
billionnaires_df_selfmade = billionnaires[[
    "industries",
    "finalWorth",
    "selfMade",
    "country",
    "gdp_country",
    "cpi_country",
    "tax_revenue_country_country"
]].copy()

# Clean 'gdp_country': remove '$', commas, convert to float, divide by 1 billion, round to 1 decimals
billionnaires_df_selfmade['gdp_country'] = (
    billionnaires_df_selfmade['gdp_country']
    .replace('[\$,]', '', regex=True)
    .astype(float)
    .div(1_000_000_000)
    .round(1)
)

# Adjust 'finalWorth': divide by 1000, round to 1 decimal
billionnaires_df_selfmade['finalWorth'] = (billionnaires_df_selfmade['finalWorth'] / 1000).round(1)

# Round 'cpi_country' down to 1 decimal place
billionnaires_df_selfmade['cpi_country'] = billionnaires_df_selfmade['cpi_country'].round(1)

billionnaires_df_selfmade = billionnaires_df_selfmade.rename(columns={
    "industries": "Industry",
    "finalWorth": "Net Worth",
    "selfMade": "Self-Made",
    "country": "Country",
    "gdp_country": "Country GDP",
    "cpi_country": "Country CPI",
    "tax_revenue_country_country": "Country Tax Revenue"
})

billionnaires_df_selfmade.head()
```

<h3>Hypothesis Testing</h3>

While the countries with the two highest GDPs show a higher percentage of self-made billionaires versus inherited wealth, is the probability of becoming a self-made billionaire increased by the GDP of the country overall?

H0: There is no statistical significance between the GDP of the country and the number of self made-billionnaires<br>
H1: The GDP of a country influences the number of self made-billionnaires  <br>


Using Pearson Correlation Coefficient<br>
Pearson's correlation coefficient, denoted by "r", is used to measure the strength and direction of a linear relationship between two continuous variables. 

```python
from scipy.stats import pearsonr

# Remove rows with missing values in 'Country GDP' or 'Self-Made'
df = billionnaires_df_selfmade.dropna(subset=['Country GDP', 'Self-Made'])

# For each country, calculate:
# - Number of self-made billionaires
# - Total number of billionaires
# - GDP
country_stats = (
    df.groupby('Country')
    .agg(
        gdp=('Country GDP', 'first'),
        self_made_count=('Self-Made', lambda x: x.sum()),
        total_count=('Self-Made', 'count')
    )
    .reset_index()
)

# Calculate proportion of self-made billionaires per country
country_stats['self_made_ratio'] = country_stats['self_made_count'] / country_stats['total_count']

# Remove countries with very low billionaire counts (optional, for stability)
country_stats = country_stats[country_stats['total_count'] >= 5]

# Pearson correlation between GDP and self-made ratio
corr, pval = pearsonr(country_stats['gdp'], country_stats['self_made_ratio'])

print(f"Pearson correlation between country GDP and self-made billionaire ratio: {corr:.3f}")
print(f"P-value: {pval:.4f}")

if pval < 0.05:
    print("There is a statistically significant relationship between GDP and the probability of a billionnaire being self-made.")
    print('H1 is accepted: The probability of a billionaire being self-made is positively correlated with GDP.')
    
else:
    print("There is no statistically significant relationship between GDP and the probability of a billionnaire being self-made.")
    print('H1 is rejected: The probability of a billionaire being self-made is not positively correlated with GDP.')


# Scatter plot of GDP vs Self-Made Ratio
plt.figure(figsize=(10, 6))
sns.scatterplot(data=country_stats, x='gdp', y='self_made_ratio', color='blue')
plt.title('GDP vs Self-Made Billionaire Ratio by Country')
plt.xlabel('Country GDP ($ Billions)')
plt.ylabel('Self-Made Billionaire Ratio')
plt.xscale('log')
plt.axhline(y=0.5, color='r', linestyle='--', label='50% Self-Made')
plt.legend()
plt.tight_layout()
plt.show()
```

The visualization above confirms there is no statistical correlation between GDP and the number of self made billionaires.The distribution is not linear nor directional.

Comments here

```python
from scipy.stats import chi2_contingency

# Create a contingency table: rows=Industry, columns=Self-Made/Inherited
contingency = industries_selfmade_inherited.set_index('Industry')[['Self-Made', 'Inherited']]

# Perform Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency)

print(f"Chi-squared statistic: {chi2:.2f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("There is a statistically significant association between industry and the distribution of self-made vs inherited billionaires.")
else:
    print("There is no statistically significant association between industry and the distribution of self-made vs inherited billionaires.")
```

This means that statistically, the choice of industry can increase the change of becoming a self-made billionnaire.

Is there a statistical relationship between the number of self-made and inherited wealth billionnaires based on industry?

H0: There is no significant statistical relationship between the number of self-made and inherited wealth billionnaires based in industries. <br>
H1: There is a significant statistical relationship between the number of self-made and inherited wealth billionnaires based in industries.

```python
from scipy.stats import f_oneway

# Get top 5 industries by number of billionaires
top5_industries = IndustryWealthCorrelation.head(5)['Industry'].tolist()

# Filter data for top 5 industries and drop rows with missing GDP or Self-Made
df_top5 = billionnaires_df_selfmade[
    (billionnaires_df_selfmade['Industry'].isin(top5_industries)) &
    (~billionnaires_df_selfmade['Country GDP'].isna())
]

# For each industry, split GDPs by self-made and inherited
results = {}
for industry in top5_industries:
    industry_df = df_top5[df_top5['Industry'] == industry]
    gdp_selfmade = industry_df[industry_df['Self-Made']]['Country GDP']
    gdp_inherited = industry_df[~industry_df['Self-Made']]['Country GDP']
    results[industry] = (gdp_selfmade, gdp_inherited)

# ANOVA for GDP of self-made vs inherited for each industry
for industry, (gdp_selfmade, gdp_inherited) in results.items():
    if len(gdp_selfmade) > 1 and len(gdp_inherited) > 1:
        f_stat, p_val = f_oneway(gdp_selfmade, gdp_inherited)
        print(f"{industry}: ANOVA F={f_stat:.2f}, p={p_val:.4f}")
        if p_val < 0.05:
            print(f"  Statistically significant difference in GDP between self-made and inherited billionaires in {industry}.")
        else:
            print(f"  No statistically significant difference in GDP between self-made and inherited billionaires in {industry}.")
    else:
        print(f"{industry}: Not enough data for statistical test.")
```
